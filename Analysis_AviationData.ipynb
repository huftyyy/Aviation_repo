```python
# plot showing mean fatal injuries per incident by aircraft category
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# Filter out 'Unknown' Aircraft Categories for risk analysis
df_analyzed = df[df['Aircraft.Category'] != 'Unknown'].copy()

# Calculate incident counts and mean fatal injuries per Aircraft Category
aircraft_category_risk = df_analyzed.groupby('Aircraft.Category').agg(
    incident_count=('Event.Id', 'count'),
    mean_fatal_injuries=('Total.Fatal.Injuries', 'mean')
).reset_index()

# Sort by mean fatal injuries for the bar plot order (highest to lowest)
plot_df = aircraft_category_risk.sort_values('mean_fatal_injuries', ascending=False)

# Create the bar plot
plt.figure(figsize=(12, 7)) # Set the figure size
# Use a colormap to get different colors for bars
colors = plt.cm.magma(np.linspace(0, 1, len(plot_df)))
plt.bar(plot_df['Aircraft.Category'], plot_df['mean_fatal_injuries'], color=colors)
plt.title('Mean Fatal Injuries per Incident by Aircraft Category') # Set the title
plt.xlabel('Aircraft Category') # Set the x-axis label
plt.ylabel('Mean Fatal Injuries') # Set the y-axis label
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.savefig('category_mean_fatal_mpl.png') # Save the plot
plt.show()
```

```python
# plot showing proprtion of 'destroyed' damage by air craft category(excluding unknown)
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Filter out 'Unknown' Aircraft Categories for risk analysis
df_analyzed = df[df['Aircraft.Category'] != 'Unknown'].copy()

# Calculate proportion of destroyed damage per Aircraft Category
aircraft_category_risk = df_analyzed.groupby('Aircraft.Category').agg(
    prop_destroyed_damage=('Aircraft.damage', lambda x: (x == 'Destroyed').sum() / len(x))
).reset_index()

# Sort by proportion of destroyed damage for the bar plot order (highest to lowest)
plot_df = aircraft_category_risk.sort_values('prop_destroyed_damage', ascending=False)

# Create the bar plot
plt.figure(figsize=(12, 7)) # Set the figure size
# Use a colormap to get different colors for bars
colors = plt.cm.plasma(np.linspace(0, 1, len(plot_df)))
plt.bar(plot_df['Aircraft.Category'], plot_df['prop_destroyed_damage'], color=colors)
plt.title('Proportion of "Destroyed" Damage by Aircraft Category') # Set the title
plt.xlabel('Aircraft Category') # Set the x-axis label
plt.ylabel('Proportion of Destroyed Incidents') # Set the y-axis label
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.savefig('category_prop_destroyed_mpl.png') # Save the plot
plt.show()
```

```python
# plot showing  Median Fatal Injuries per Incident for Top Airplane Make/Models

import matplotlib.pyplot as plt
import numpy as np
# Group by Make and Model and calculate the median of 'Total.Fatal.Injuries'
fatal_injuries_by_model = df.groupby(['Make', 'Model'])['Total.Fatal.Injuries'].median().reset_index()

# Sort by the median fatal injuries in descending order
fatal_injuries_by_model_sorted = fatal_injuries_by_model.sort_values('Total.Fatal.Injuries', ascending=False)

# Select the top N models (e.g., top 10)
top_n = 20
top_models = fatal_injuries_by_model_sorted.head(top_n)

# Create a combined 'Make - Model' string for plotting
top_models['Make - Model'] = top_models['Make'] + ' - ' + top_models['Model']

# Create the plot
plt.figure(figsize=(12, 8))
plt.bar(top_models['Make - Model'], top_models['Total.Fatal.Injuries'], color=plt.cm.viridis(np.linspace(0, 1, top_n)))
plt.title(f'Top {top_n} Airplane Make/Models by Median Fatal Injuries per Incident')
plt.xlabel('Make - Model')
plt.ylabel('Median Total Fatal Injuries')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('median_fatal_injuries_top_models_mpl.png')
plt.show();
```

```python
# plot showing Average Fatal Injuries per Incident for the Top 20 Most Common Airplane Make/Models.

import matplotlib.pyplot as plt
import numpy as np
# Combine Make and Model into a single column
df['Make_Model'] = df['Make'] + ' ' + df['Model']

# Calculate the number of incidents per Make/Model
model_incident_counts = df['Make_Model'].value_counts()

# Get the top 20 most common Make/Models
top_20_models = model_incident_counts.head(20).index

# Filter the DataFrame to include only the top 20 models
df_top_models = df[df['Make_Model'].isin(top_20_models)].copy()

# Group by the combined 'Make_Model' and calculate the average fatal injuries
average_fatal_injuries = df_top_models.groupby('Make_Model')['Total.Fatal.Injuries'].mean().sort_values(ascending=False)

# Create the bar plot
plt.figure(figsize=(14, 8))
average_fatal_injuries.plot(kind='bar', color=plt.cm.tab20(np.arange(len(average_fatal_injuries))))
plt.title('Average Fatal Injuries per Incident for the Top 20 Most Common Airplane Make/Models')
plt.xlabel('Airplane Make/Model')
plt.ylabel('Average Fatal Injuries')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.savefig('average_fatal_injuries_top_20_models_mpl.png')
plt.show()
```

```python
# plot showing Proportion of "Destroyed" Damage for Top Airplane Make/Models

import matplotlib.pyplot as plt
import numpy as np
# Filter for incidents where Aircraft.damage is 'Destroyed'
destroyed_incidents = df[df['Aircraft.damage'] == 'Destroyed'].copy()

# Combine Make and Model into a single column for the destroyed incidents
destroyed_incidents['Make_Model'] = destroyed_incidents['Make'] + ' ' + destroyed_incidents['Model']

# Calculate the number of destroyed incidents per Make/Model for the top models
destroyed_counts_top_models = destroyed_incidents[destroyed_incidents['Make_Model'].isin(top_20_models)]['Make_Model'].value_counts()

# Calculate the total number of incidents per Make/Model for the top models
total_counts_top_models = df_top_models['Make_Model'].value_counts()

# Calculate the proportion of destroyed damage for each of the top 20 models
# Ensure we handle cases where a model might not have any destroyed incidents
proportion_destroyed = destroyed_counts_top_models.reindex(top_20_models).fillna(0) / total_counts_top_models.reindex(top_20_models).fillna(0)

# Sort the proportion in descending order
proportion_destroyed_sorted = proportion_destroyed.sort_values(ascending=False)

# Create the bar plot
plt.figure(figsize=(14, 8))
proportion_destroyed_sorted.plot(kind='bar', color=plt.cm.viridis(np.arange(len(proportion_destroyed_sorted))))
plt.title('Proportion of "Destroyed" Damage for the Top 20 Most Common Airplane Make/Models')
plt.xlabel('Airplane Make/Model')
plt.ylabel('Proportion of Incidents Resulting in "Destroyed" Damage')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.savefig('proportion_destroyed_damage_top_20_models_mpl.png')
plt.show()
```

```python
# plot showing Accident rates over time (1962â€“2023) by Aircraft.Category


import pandas as pd
import matplotlib.pyplot as plt
# Accident rates over time (1962â€“2023) by Aircraft.Category, showing trends toward safer modern aircraft  excluding unknown
# Filter out 'unknown' Aircraft.Category
df_filtered = df[df['Aircraft.Category'] != 'unknown'].copy()
# Convert 'Event.Date' to datetime and extract the year
df_filtered['Event.Year'] = pd.to_datetime(df_filtered['Event.Date']).dt.year
# Group by year and aircraft category, then count incidents
yearly_category_incidents = df_filtered.groupby(['Event.Year', 'Aircraft.Category']).size().unstack(fill_value=0)
# Filter years from 1962 to 2023
yearly_category_incidents = yearly_category_incidents.loc[1962:2023]
# Plotting
plt.figure(figsize=(15, 8))
for category in yearly_category_incidents.columns:
    plt.plot(yearly_category_incidents.index, yearly_category_incidents[category], label=category)
plt.title('Accident Rates Over Time (1962â€“2023) by Aircraft Category')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.legend(title='Aircraft Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('accident_rates_over_time_by_category_mpl.png')
plt.show()
```
